```python
import numpy as np
import pandas as pd
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
#genres
#id
#keywords
#title
#overview
#cast
#crew

movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head()

```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['keywords']=movies['keywords'].apply(convert)
movies['genres']=movies['genres'].apply(convert)
movies.head()
```

```python
def extract_cast_names(cast_str):
    cast_list = ast.literal_eval(cast_str)
    return [member['name'] for member in cast_list[:3]]

movies['cast'] = movies['cast'].apply(extract_cast_names)
movies.head()
```

```python
def extract_director(crew_str):
    crew_list = ast.literal_eval(crew_str)
    return [member['name'] for member in crew_list if member.get('job') == 'Director']

movies['crew'] = movies['crew'].apply(extract_director)
movies.head()
```

```python
movies['overview'] = movies['overview'].astype(str).apply(lambda x: x.split())
movies.head()
```

```python
for col in ['genres', 'keywords', 'cast', 'crew']:
    movies[col] = movies[col].apply(lambda lst: [x.replace(' ', '') for x in lst] if isinstance(lst, list) else lst)

movies.head()
```

```python
movies['tag'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.drop(columns=['overview', 'genres', 'keywords', 'cast', 'crew'], inplace=True)
movies.head()
```

```python
movies['tag'] = movies['tag'].apply(lambda x: " ".join(x) if isinstance(x, list) else str(x))
movies.head()
```

```python
movies['tag'] = movies['tag'].str.lower()
```

```python
movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vectors = cv.fit_transform(movies['tag']).toarray()
vectors
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies['tag'] = movies['tag'].apply(stem)
movies.head()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity
```

```python
def recommend(movie):
    movie_index = movies[movies['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(movies.iloc[i[0]].title)
recommend('Avatar')
```

```python
import pickle as pkl
pkl.dump(movies.to_dict(),open('movies_dict.pkl','wb'))
pkl.dump(similarity,open('similarity.pkl','wb'))
```
